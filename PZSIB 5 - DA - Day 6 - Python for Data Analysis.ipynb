# Data Preprocessing

```python
import pandas as pd
```

## Merging data

### Joins

```python
df1 = pd.DataFrame({
    'id'    : [1, 2, 3, 4],
    'name'  : ['Budi', 'Yosua', 'Adel', 'Dinda']
})

df1
```

```python
df2 = pd.DataFrame({
    'id'    : [1, 2, 3, 5, 6],
    'age'  : [20, 30, 40, 20, 25]
})

df2
```

```python
df_inner_join = pd.merge(df1, df2, on = 'id')
df_inner_join
```

```python
df_left_join = pd.merge(df1, df2, on='id', how='left')
df_left_join
```

```python
df_right_join = pd.merge(df1, df2, on='id', how='right')
df_right_join
```

```python
df_full_outer = pd.merge(df1, df2, on='id', how='outer')
```

```python
df_left_d2 = pd.merge(df2, df1, on='id', how='left')
df_left_d2
```

## Data Aggregations

```python
df = pd.DataFrame({
    'Animal':['Cheetah','Cheetah','Eagle','Eagle','Eagle'],
    'Speed':[80,85,100,110,80]
    })
df
```

```python
df.groupby('Animal').sum()
```

```python
df.groupby('Animal').count()
```

```python
df.groupby('Animal').describe()
```

```python
df.groupby('Animal').agg({
    'Speed': ['mean', 'min', 'max']
})
```
