### Data Encoding

```python
df = pd.DataFrame({
    'nama'    : ['A', 'B', 'C', 'D'],
    'nilai'  : [90,95,87,75],
    'survey'  : ['suka', 'biasa', 'tidak suka', 'suka']
})

df
```

```python
# add column and value
df['kelas'] = 'Data Analytics'
#df['kelas'] = 3 <-- kalau datanya numeric
df
```

```python
# change value
df.loc[(df['kelas'] == 'Data Analytics'), 'kelas'] = 'Studi Independen'
```

```python
df.loc[(df['nilai'] == 90), 'kelas'] = np.nan
```

```python
df_no_90 = df[df['nilai']!=90]
df_no_90
```

```python
#df['nilai'] = df['nilai'].astype('str')
df['nilai'] = df['nilai'].astype('float64')

df.dtypes
```

### Anomalies and Outlier Handling

```python
#Impor dataset dari library lain, contohnya seaborn
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['fare'])
plt.show()
```

```python
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
outliers = df[df['fare'] > upper_bound]
outliers
```
