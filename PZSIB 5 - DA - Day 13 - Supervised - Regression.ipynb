# Supervised Learning

## Import

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
```

## Load Dataset

```python
# Import data dari CSV ke dataframe
df = pd.read_csv('https://raw.githubusercontent.com/iamontheinet/datascience/master/Databricks_ML/Advertising.csv')
df.head()
```

```python
data = df[['TV','Sales']]
data.head(3)
```

## Simple Linear Regression

```python
#Plot the data
fig, ax = plt.subplots(figsize=(7,5))
sns.scatterplot(x='TV', y='Sales', data=data, s=70)

#Menambahkan title
title = 'Sales and TV'
ax.text(45,26,title,horizontalalignment='center',color='#800000',fontsize=20,fontweight='bold')

#Menambahkan judul sumbu x dan sumbu y
ax.set_xlabel('Television', fontsize=15, color='#800000')
ax.set_ylabel('Sales', fontsize=14, color='#800000')
```

```python
#Mengganti nama kolom menjadi x dan y
data.columns = ['x','y']
data.head()
```

## Simple Linear Regression from Scratch

```python
#Rumus Regresi Linear secara matematis
data['(x-x_mean)'] = round(data['x'] - data['x'].mean(),2)
data['(y-y_mean)'] = round(data['y'] - data['y'].mean(),2)
data['(x-x_mean)(y-y_mean)'] = data['(x-x_mean)'] * data['(y-y_mean)']
data['pow((x-x_mean),2)'] = round(pow(data['x'] - data['x'].mean(),2),2)
data.head()
```

```python
#Rumus menghitung slope (b1)
b1 = data['(x-x_mean)(y-y_mean)'].sum() / data['pow((x-x_mean),2)'].sum()
b1
```

```python
#Menghitung rata-rata y dan x
data['y'].mean(), data['x'].mean()
```

```python
#Rumus menghitung intercept (b0)
b0 = data['y'].mean() - b1 * data['x'].mean()
b0
```

```python
textstr = 'y = '+str(b0)+' + '+str(b1)+' * x'
textstr
```

```python
data['y_pred'] = b0 + b1*data['x']
data.head(5)
```

## Simple Linear Regression with sklearn

```python
# Membuat data menjadi array sesuai dengan kebutuhan model
x = data.x.values.reshape(-1,1)
y = data.y.values.reshape(-1,1)

#Model Linear Regression
model = LinearRegression().fit(x, y)

#Menghitung b0 dan b1
b0 = float(model.intercept_[0])
b0 = round(b0,3)
b1 = float(model.coef_[0][0])
b1 = round(b1,3)

#Menampilkan hasil
print('intercept (b0 ): ', b0)
print('slope (b1) : ', b1)
print(' ')
textstr = 'y = '+str(b0)+' + '+str(b1)+' * x'
print(textstr)
```

```python
fit = np.polyfit(data['x'], data['y'], deg=1)
y_hat = fit[1] + data['x'] * fit[0]

fig, ax = plt.subplots(figsize=(10,7))
sns.regplot(x='x', y='y', data=data)
plt.vlines(data['x'], y_hat, data['y'], lw = .4, colors='r');


#Menambahkan title
title = 'TV and Sales'
ax.text(1,26,title,horizontalalignment='left',color='#800000',fontsize=20,fontweight='bold')
ax.text(46,24,textstr,horizontalalignment='center',color='#800000',fontsize=14,fontweight='bold')
ax.set_xlabel('TV', fontsize=15, color='#800000')
ax.set_ylabel('Sales', fontsize=14, color='#800000')
```
