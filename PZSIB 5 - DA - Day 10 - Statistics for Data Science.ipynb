# Import Packages

```python
import pandas as pd
import scipy.stats as stats
alpha = 0.05
```

# First Problem

## Assumption Check

```python
# Define the data and alpha
data_1 = [11.8, 8.6, 12.6, 7.9, 6.4, 10.4, 13.6, 9.1]
stats.shapiro(data_1)
stat, p = stats.shapiro(data_1)
print(p)
```

```python
# Shapiro-Wilk test to check normality of a data
# p_SW_1 is variable name for p-value. This name is optional
_, p_SW_1 = stats.shapiro(data_1)
p_SW_1 = round(p_SW_1,2) #membulatkan p_value ke 2 desimal

# Check the p-value. If it's bigger than alpha, then the data is normally distributed
if p_SW_1 > alpha:
    print('Data appears to be normally distributed (p-value =', p_SW_1, ')')
else:
    print('Data does not appear to be normally distributed (p-value =', p_SW_1, ')')
```

## One sample t-test

```python
stats.ttest_1samp(data_1, 11)
```

```python
# Perform the one-sample t-test using ttest_1samp function from stats
_, p_value_1 = stats.ttest_1samp(data_1, 11)
p_value_1 = round(p_value_1,2)
print('p-value:', p_value_1)

# Check the p-value to determine significance. If p-value is smaller than alpha, reject null hypothesis
if p_value_1 < alpha:
    print('Because',p_value_1,'< 0.05, we reject the null hypothesis')
else:
    print('Because',p_value_1,'> 0.05, we accept the null hypothesis')
```

```python
# alternative is used when the hypothesis includes > or <
_, p_value_1 = stats.ttest_1samp(data_1, 11, alternative = "less")
p_value_1
```

# Second Problem

```python
disaster_count_df = pd.read_csv('https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/number-of-natural-disaster-events.csv')
disaster_count_df.rename(columns={'Number of reported natural disasters (reported disasters)': 'Disaster Count'}, inplace = True)
disaster_count_df = disaster_count_df[(disaster_count_df['Year'] >= 2001) & (disaster_count_df['Year'] <= 2010) & (disaster_count_df['Entity'] == 'All natural disasters')]
disaster_count_df
```

## Assumption Check

```python
# Shapiro-Wilk test
_, p_SW_2 = stats.shapiro(disaster_count_df['Disaster Count'])

if p_SW_2 > alpha:
    print('Data appears to be normally distributed (p-value =', p_SW_2, ')')
else:
    print('Data does not appear to be normally distributed (p-value =', p_SW_2, ')')
```

## One sample t-test

```python
# Perform the one-sample t-test
_, p_value_2 = stats.ttest_1samp(disaster_count_df['Disaster Count'], 400)
p_value_2 = round(p_value_2,2)
print('p-value:', p_value_2)

# Check the p-value to determine significance
if p_value_2 < alpha:
    print('Because',p_value_2,'< 0.05, we reject the null hypothesis')
else:
    print('Because',p_value_2,'> 0.05, we accept the null hypothesis')
```

# Third Problem

```python
# Pull the data from github
economic_damage_df = pd.read_csv('https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv')
# Rename column name that's too long
economic_damage_df.rename(columns={'Total economic damage from natural disasters (US$)': 'Economic Damage'}, inplace = True)
# The first dataframe is economic damage from 2001 to 2010, Flood only
economic_damage_df_1 = economic_damage_df[(economic_damage_df['Year'] >= 2001) & (economic_damage_df['Year'] <= 2010) & (economic_damage_df['Entity'] == 'Flood')]
# The first dataframe is economic damage from 2011 to 2020, Flood only
economic_damage_df_2 = economic_damage_df[(economic_damage_df['Year'] >= 2011) & (economic_damage_df['Year'] <= 2020) & (economic_damage_df['Entity'] == 'Flood')]
```

```python
economic_damage_df_1
```

```python
economic_damage_df_2
```

## Assumption Check

```python
# Shapiro-Wilk test
_, p_SW_3_economic_damage_1 = stats.shapiro(economic_damage_df_1['Economic Damage'])
_, p_SW_3_economic_damage_2 = stats.shapiro(economic_damage_df_2['Economic Damage'])

if p_SW_3_economic_damage_1 > alpha:
    print('Economic damage data 1 appears to be normally distributed (p-value =', p_SW_3_economic_damage_1, ')')
else:
    print('Economic damage data 1 does not appear to be normally distributed (p-value =', p_SW_3_economic_damage_1, ')')

if p_SW_3_economic_damage_2 > alpha:
    print('Economic damage data 2 appears to be normally distributed (p-value =', p_SW_3_economic_damage_2, ')')
else:
    print('Economic damage data 2 does not appear to be normally distributed (p-value =', p_SW_3_economic_damage_2, ')')

# Levene's
_, p_Levene_3 = stats.levene(economic_damage_df_1['Economic Damage'], economic_damage_df_2['Economic Damage'])

if p_Levene_3 > alpha:
    print("The variances of the two datasets are homogenous (p-value =", p_Levene_3, ")")
else:
    print("The variances of the two datasets are not homogenous (p-value =", p_Levene_3, ")")
```

## Two Sample t-test

```python
# Two samples t-test
_, p_value_3 = stats.ttest_ind(economic_damage_df_1['Economic Damage'], economic_damage_df_2['Economic Damage'], alternative = 'less')
p_value_3 = round (p_value_3,2)

# Check the p-value to determine significance
if p_value_3 < alpha:
    print('Because',p_value_3,'< 0.05, we reject the null hypothesis')
else:
    print('Because',p_value_3,'> 0.05, we accept the null hypothesis')
```
