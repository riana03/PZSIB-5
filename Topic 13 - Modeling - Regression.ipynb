```python
import pandas as pd
import numpy as np
import warnings
```

```python
# Set max display rows & columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

warnings.filterwarnings("ignore")
```

```python
ds = pd.read_csv('used_car.csv')

# https://www.kaggle.com/code/iabhishekmaurya/used-car-price-prediction/data?select=train-data.csv
# https://www.kaggle.com/code/iabhishekmaurya/used-car-price-prediction/data?select=test-data.csv
```

```python
ds.head()
```

```python
ds = ds.drop(['Unnamed: 0'],1)
```

```python
ds.dtypes
```

```python
ds.isnull().sum()
```

```python
ds = ds.drop(['New_Price', 'Name'],1)
```

```python
ds = ds.dropna()
```

#### One Hot Encoding

```python
ds =  pd.get_dummies(ds, columns=['Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage', 
                                     'Engine', 'Power'], drop_first=True)
ds.head()
```

```python
data = ds.drop(['Price'], 1)
target = ds[['Price']]
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data
Y = target
train_X, test_X, train_y, test_y = train_test_split(X,Y,test_size=0.2, random_state= 0)
```

#### Model Evaluation

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

#### Linier Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(train_X, train_y)
linregpred = linreg.predict(test_X)
```

```python
# MAE
mean_absolute_error(test_y, linreg.predict(test_X))
```

```python
# MSE
mean_squared_error(test_y, linreg.predict(test_X))
```

```python
# RMSE
mean_squared_error(test_y, linreg.predict(test_X), squared=False)
```

```python
# MAPE
mean_absolute_percentage_error(test_y, linreg.predict(test_X))
```

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decisiontree = DecisionTreeRegressor()
decisiontree.fit(train_X, train_y)
dtpred = decisiontree.predict(test_X)
```

```python
# MAE
mean_absolute_error(test_y, decisiontree.predict(test_X))
```

```python
# MSE
mean_squared_error(test_y, decisiontree.predict(test_X))
```

```python
# RMSE
mean_squared_error(test_y, decisiontree.predict(test_X), squared=False)
```

```python
# MAPE
mean_absolute_percentage_error(test_y, decisiontree.predict(test_X))
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
rf.fit(train_X, train_y)
rfpred = rf.predict(test_X)
```

```python
# MAE
mean_absolute_error(test_y, rf.predict(test_X))
```

```python
# MSE
mean_squared_error(test_y, rf.predict(test_X))
```

```python
# RMSE
mean_squared_error(test_y, rf.predict(test_X), squared=False)
```

```python
# MAPE
mean_absolute_percentage_error(test_y, rf.predict(test_X))
```
