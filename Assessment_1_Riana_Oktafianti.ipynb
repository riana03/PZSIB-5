```python
# Import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as pl
```

```python
# Membaca file csv yang  digunakan

data = '/content/Exploratory-data-analysis-of-climate-change-and-natural-disasters/datasets/number-of-natural-disaster-events.csv'
df = pd.read_csv(data)
df
```

```python
# Memisahkan variabel numerik dan variabel categorical
categorical_variables = df.iloc[:, 0]
numerical_variables = df.iloc[:, 2]
numerical_variables_1 = df.iloc[:, 3]

# Menampilkan beberapa baris pertama dari masing-masing variabel
print("Variabel Kategorikal:")
print(categorical_variables.head())

print("\nVariabel Numerik:")
print(numerical_variables.head())

print("\nVariabel Numerik:")
print(numerical_variables_1.head())
```

**Barplot Vertikal**

```python
# Data untuk membuat barplot
data_barplot = categorical_variables # Entity
data_barplot_1 = numerical_variables #Year
data_barplot_2 = numerical_variables_1 # Number of reported natural disasters


# Membuat barplot vertikal
plt.figure(figsize = (10, 6))
plt.bar(data_barplot, data_barplot_2, color='c')
plt.title('Kejadian Bencana Alam Tahun 1900 - 2018 ', size = 16)
plt.xlabel('Jenis Bencana Alam', size = 12)
plt.ylabel('Total kejadian', size = 12)
plt.xticks(rotation = 60)

# Menampilkan plot
plt.show()
```

```python
# Membuat barplot vertikal
plt.figure(figsize = (20, 6))
plt.bar(data_barplot_1, data_barplot_2, color='brown')
plt.title('Kejadian Bencana Alam Tahun 1900 - 2018 ', size = 16)
plt.xlabel('Tahun terjadi', size = 12)
plt.ylabel('Total bencana alam', size = 12)
plt.xticks(rotation = 60)

# Menampilkan plot
plt.show()
```

**Barplot Horizontal**

```python
# Membuat barplot horizontal
plt.figure(figsize=(10, 6))
plt.barh(data_barplot, data_barplot_2, color='m')
plt.title('Kejadian Bencana Alam Tahun 1900 - 2018 ', size = 16)
plt.ylabel('Jenis Bencana Alam', size = 12)
plt.xlabel('Total kejadian', size = 12)

# Menampilkan plot
plt.show()
```

```python
# Membuat barplot horizontal
plt.figure(figsize=(10, 12))
plt.barh(data_barplot_1, data_barplot_2, color = 'red')
plt.title('Kejadian Bencana Alam Tahun 1900 - 2018 ', size = 16)
plt.xlabel('Jumlah bencana alam', size = 12)
plt.ylabel('Tahun terjadi', size = 12)

# Menampilkan plot
plt.show()
```

**Distplot**

```python
# Data untuk membuat distplot
data_distplot = categorical_variables # Entity
data_barplot_1 = numerical_variables #Year
data_barplot_2 = numerical_variables_1 # Number of reported natural disasters
```

```python
# Membuat distplot
plt.figure(figsize = (10, 6))
sns.histplot(data_distplot, kde = True, color = 'c')
plt.title('Distribusi Data Bencana Alam', size = 16)
plt.xlabel('Nilai Numerik', size = 12)
plt.ylabel('Frekuensi', size = 12)
plt.xticks(rotation = 60)

# Menampilkan plot
plt.show()
```
