# Preparation

```python
# Import module
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Collection

```python
df = pd.read_csv('https://raw.githubusercontent.com/andrewcole33/telco_churn_analysis/master/data/churn_data.csv')
df.head()
```

```python
df.columns
```

# Data Cleansing

```python
df = df.rename(columns = {
    'Entity' : 'entity',
    'Code' : 'code',
    'Year' : 'year',
    'Number of reported natural disasters (reported disasters)' : 'cases'
})
df.head(2)
```

```python
#Melihat statistika deskriptif dari data numerik
df.describe()
```

```python
# izin bertanya tuan, untuk kolom code kan semuanya bernilai NaN, bagaimana baiknya, apakah dibuang saja code nya atau dibiarkan saja
df = df.drop('code', axis=1)
```

```python
# Melihat data unik dari data kategorik
df.entity.unique()
```

## Entity: All Natural Disasters

```python
df_all = df[df['entity'] == 'All natural disasters']
df_all.head(2)
```

```python
df_drought = df[df['entity']=='Drought']
df_drought.head(2)
```

## Add Title
